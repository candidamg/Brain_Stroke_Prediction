```python
# Initial imports
import pandas as pd
from pathlib import Path
from sklearn import tree
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# Loading data
file_path = Path("Resources/clean_data.csv")
df1 = pd.read_csv(file_path)
df1.head()
```

```python
#Drop the id column caus it doesn't add any information
df1.drop(columns=['id'], inplace=True)
```

```python
numeric= pd.get_dummies(df1)
numeric
```

```python
# Define features set
X = numeric.copy()
X.drop("stroke", axis=1, inplace=True)
X.head()
```

```python
# Define target vector
y = numeric["stroke"].values.reshape(-1, 1)
y[:5]
```

```python
# Splitting into Train and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, train_size=0.50)
```

```python
# Creating StandardScaler instance
scaler = StandardScaler()
```

```python
# Fitting Standard Scaller
X_scaler = scaler.fit(X_train)
```

```python
# Scaling data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Creating the decision tree classifier instance
model = tree.DecisionTreeClassifier()
```

```python
# Fitting the model
model = model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data
predictions = model.predict(X_test_scaled)
```

```python
# Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
)

# Calculating the accuracy score
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
m_label=("Decision Tree","Decision Tree")
m_accuracy=(acc_score,acc_score)
cm_df["Model"] = m_label
cm_df["Accuracy"] = m_accuracy

cm_df2=cm_df.reset_index()

cm_df2.to_csv("./Resources/Model_Results.csv",header=False,mode="a" )
```

# Oversampling

```python
# Import the RandomOverSampler module form imbalanced-learn
from imblearn.over_sampling import RandomOverSampler

# Instantiate the random oversampler model
# # Assign a random_state parameter of 1 to the model
ros = RandomOverSampler(random_state=1)

# Fit the original training data to the random_oversampler model
X_res, y_res = ros.fit_resample(X, y)
```

```python
# Count the distinct values of the resampled labels data
pd.value_counts(y_res)
```

```python
# Fitting Standard Scaller
#X_scaler = scaler.fit(X_res)
# Scaling data
X_train_scaled = X_scaler.transform(X_res)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Creating the Random Forest classifier instance
rf_model =tree.DecisionTreeClassifier()
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_res)
# Making predictions using the testing data
predictions = rf_model.predict(X_test_scaled)
```

```python
# Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
)

# Calculating the accuracy score
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
m_label=("Oversampled Tree","Oversampled Tree")
m_accuracy=(acc_score,acc_score)
cm_df["Model"] = m_label
cm_df["Accuracy"] = m_accuracy

cm_df2=cm_df.reset_index()

cm_df2.to_csv("./Resources/Model_Results.csv",header=False,mode="a" )
```
